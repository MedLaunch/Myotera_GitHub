```python
import json
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sb
```

```python
acc = []
with open('jumping_jacks_52Hz.json','r') as f:
    for line in f:
        acc.append(json.loads(line))
print(acc[0])
len(acc)
```

```python
it = 0
for i in acc:
    print(i.get('acc').keys(), '\n')
    print(i.get('acc').get('ArrayAcc'))
    it += 1
    if it > 1:
        break
```

```python
timestamps = []
xF = []
x1 = []
x2 = []
x3 = []
yF = []
y1 = []
y2 = []
y3 = []
zF = []
z1 = []
z2 = []
z3 = []
```

```python
for entry in acc:
    stamp = entry.get('acc').get('Timestamp')
    
    x_F = entry.get('acc').get('ArrayAcc')[0].get('x')
    y_F = entry.get('acc').get('ArrayAcc')[0].get('y')
    z_F = entry.get('acc').get('ArrayAcc')[0].get('z')
    
    x_1 = entry.get('acc').get('ArrayAcc')[1].get('x')
    y_1 = entry.get('acc').get('ArrayAcc')[1].get('y')
    z_1 = entry.get('acc').get('ArrayAcc')[1].get('z')
    
    x_2 = entry.get('acc').get('ArrayAcc')[2].get('x')
    y_2 = entry.get('acc').get('ArrayAcc')[2].get('y')
    z_2 = entry.get('acc').get('ArrayAcc')[2].get('z')
    
    x_3 = entry.get('acc').get('ArrayAcc')[3].get('x')
    y_3 = entry.get('acc').get('ArrayAcc')[3].get('y')
    z_3 = entry.get('acc').get('ArrayAcc')[3].get('z')
    
    timestamps.append(stamp)
    
    xF.append(x_F)
    yF.append(y_F)
    zF.append(z_F)
    
    x1.append(x_1)
    y1.append(y_1)
    z1.append(z_1)
    
    x2.append(x_2)
    y2.append(y_2)
    z2.append(z_2)
    
    x3.append(x_3)
    y3.append(y_3)
    z3.append(z_3)
```

```python
x_components = np.array([x1,x2,x3]).T
y_components = np.array([y1,y2,y3]).T
z_components = np.array([z1,z2,z3]).T

fused_signals = np.array([xF, yF, zF])
df = pd.DataFrame(fused_signals.T,columns = ['x','y','z'])
df
```
